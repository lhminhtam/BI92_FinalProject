## RFM

```python
RFM = df.groupby('Customer ID').agg(Max_InvoiceDate = ("Order Date", "max"), F = ("Order ID", "nunique"), M = ("Sales", "sum")).reset_index()
```

```python
RFM['Today'] = pd.to_datetime("2025-07-26")
```

```python
RFM["R"] = (RFM["Today"] - RFM["Max_InvoiceDate"]).dt.days
```

```python
RFM = RFM[["Customer ID", "R", "F", "M"]]
RFM
```

```python
RFM['F_Score'] = pd.qcut(RFM['F'].rank(method = 'first'), 5, labels = [1, 2, 3, 4, 5])
RFM['M_Score'] = pd.qcut(RFM['M'].rank(method = 'first'), 5, labels = [1, 2, 3, 4, 5])
RFM['R_Score'] = pd.qcut(RFM['R'].rank(method = 'first'), 5, labels = [5, 4, 3, 2, 1])
```

```python
RFM['RFM_Score'] = RFM['R_Score'].astype(str) + RFM['F_Score'].astype(str) + RFM['M_Score'].astype(str)
```

```python
rfm_ranking = pd.read_csv('/content/RankRFM.xlsx - Sheet1.csv')
```

```python
rfm_ranking['Scores'] = rfm_ranking['Scores'].str.replace(', ', ',').str.split(",")
```

```python
rfm_ranking_new = rfm_ranking.explode("Scores").reset_index()
```

```python
rfm_ranking_new
```

```python
RFM = RFM.merge(rfm_ranking_new, how = "left", left_on = "RFM_Score", right_on="Scores")
RFM
```

```python
RFM = RFM.drop(columns = ["index"])
```
